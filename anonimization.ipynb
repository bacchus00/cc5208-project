```python
df = pd.read_csv(root+'ingresos-campanas-politicas.csv')
```

```python
# applying md5 to ruts (except to parties and candidates)

ruts_p = list(df[~df.RUTPartido.isna()].RUTPartido.unique())
ruts_c = list(df[~df.RUTCandidato.isna()].RUTCandidato.unique())

def aux_fun_md5(x):
  if (x == x) and (x not in (ruts_p+ruts_c)):
    return hashlib.md5(str(x).encode()).hexdigest()
  else:
    return x

df['RUTAportante'] = df['RUTAportante'].apply(aux_fun_md5)
```
